```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
cars = pd.read_csv('cars.csv')
cars
```

```python
cars.dtypes
```

```python
cars.isna().sum()
```

```python
#     Создайте новый признак – марку автомобиля (company). Машины каких производителей встречаются в датасете? 
#     Далее исправьте названия и проверьте изменения.
```

```python
cars['company'] = cars.CarName.apply(lambda x: x.split()[0])
```

```python
cars = cars.drop(['CarName', 'car_ID'], axis=1)
```

```python
cars
```

```python
cars.company.nunique()
```

```python
cars.company.value_counts()
```

```python
# cars.company.apply(lambda x: x.lower()) \
#     .apply(lambda x: x.replace('maxda', 'mazda')) \
#     .apply(lambda x: x.replace('toyouta', 'toyota')) \
#     .apply(lambda x: x.replace('vokswagen', 'volkswagen')) \
#     .apply(lambda x: x.replace('vw', 'volkswagen')) \
#     .apply(lambda x: x.replace('porcshce', 'porsche')) \
#     .nunique()
```

```python
cars['company'] = cars.company.str.lower() \
    .replace('maxda', 'mazda') \
    .replace('toyouta', 'toyota') \
    .replace('vokswagen', 'volkswagen') \
    .replace('vw', 'volkswagen') \
    .replace('porcshce', 'porsche') 
```

```python
cars.company.nunique()
```

```python
#     Оставьте только часть предикторов, после чего посчитайте корреляцию между price и другими переменными.
```

```python
cars[['company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 
      'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 'price']].corr() \
    .round(2)
```

```python
#     Преобразуйте категориальные переменные с помощью pd.get_dummies(). 
```

```python
cars.dtypes.sort_values()
```

```python
df_dummy = pd.get_dummies(data=cars[['enginetype', 'cylindernumber', 'enginelocation', 'drivewheel', 'carbody',
                                     'doornumber', 'aspiration', 'fueltype' , 'fuelsystem', 'company']], drop_first = True)
```

```python
df_dummy
```

```python
df_dummy_2 = pd.get_dummies(data=cars[['company', 'fueltype', 'aspiration','carbody', 'drivewheel', 
                                       'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 
                                       'cylindernumber', 'enginesize', 'boreratio','horsepower', 'price']], drop_first = True)
```

```python
df_dummy_2
```

```python
#     Постройте модель с одним предиктором цены – horsepower. Какой процент изменчивости объясняет 
#     полученная модель? (\(R^2\))
```

```python
model_one = smf.ols('price ~ horsepower', df_dummy_2).fit().summary()
print(model_one)
```

```python
R-squared: 0.653 # 65%
```

```python
#     Далее – две модели (со всеми предикторами и со всеми, кроме марок машин). Обратите внимание на 
#     изменения в \(R^2\), коэффициентах и их значимости. Какую модель лучше оставить? 
```

```python
X1 = sm.add_constant(df_dummy_2.drop('price', axis=1))  #  модель со всеми предиктами
```

```python
model_all = sm.OLS(df_dummy_2.price, X1).fit().summary()
print(model_all)
```

```python
predicts_no_avto = df_dummy_2[['wheelbase', 'carlength', 'carwidth', 'curbweight', 'enginesize',
       'boreratio', 'horsepower', 'fueltype_gas',
       'aspiration_turbo', 'carbody_hardtop', 'carbody_hatchback',
       'carbody_sedan', 'carbody_wagon', 'drivewheel_fwd', 'drivewheel_rwd',
       'enginetype_dohcv', 'enginetype_l', 'enginetype_ohc', 'enginetype_ohcf',
       'enginetype_ohcv', 'enginetype_rotor', 'cylindernumber_five',
       'cylindernumber_four', 'cylindernumber_six', 'cylindernumber_three',
       'cylindernumber_twelve', 'cylindernumber_two']]
```

```python
X2 = sm.add_constant(predicts_no_avto)  # со всеми, кроме марок машин
```

```python
model_no_avto = sm.OLS(df_dummy_2.price, X2).fit().summary()
print(model_no_avto)
```

```python
# Вывод:
# Большинство коэффициентов, связанных с марками машин, статистически незначимы
# Если судить чисто по диагностическим показателям (вроде R2), то модель со всеми предикторами лучшая
```
