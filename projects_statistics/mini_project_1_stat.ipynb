```python
    Изучите объемы продаж игр Nintendo в Японии с разбивкой жанрам. Посмотрите на медиану, среднее, межквартильный размах.
```

```python
games_nindendo = games.query("Publisher == 'Nintendo'")[['Genre', 'JP_Sales']]
games_nindendo
```

```python
plt.figure(figsize=(16,10))
sns.boxplot(x='Genre', y='JP_Sales', data=games_nindendo)
```

```python
 Визуализируйте динамику изменения объема мировых продаж (Global_Sales) по годам для игр Nintendo следующих жанров: 
        Fighting, Simulation, Platform, Racing, Sports. В какие моменты времени было продано больше всего игр жанра Sports?
```

```python
nindendo = games.query("Publisher == 'Nintendo' and \
                      (Genre == 'Fighting' or Genre == 'Simulation' or Genre == 'Platform' \
                      or Genre == 'Racing' or Genre == 'Sports')") \
                      [['Genre', 'Year', 'Global_Sales']]  
nindendo
```

```python
nindendo_genre = nindendo \
    .groupby(['Genre', 'Year'], as_index=False) \
    .agg({'Global_Sales': 'sum'})
nindendo_genre
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x='Year', y='Global_Sales', hue='Genre', data=nindendo_genre)
```
