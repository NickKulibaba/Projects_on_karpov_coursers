```python
# Дополнительный проект

# В папке subsid (shared/homeworks/python_ds_miniprojects/5_subsid) находятся файлы (tm_sales_1, tm_sales_2, ...) 
# с продажами продуктов через телемаркетинг. Каждый файл содержит, как минимум, 4 колонки (поля): 
#     FILIAL_ID, SUBS_ID, PROD_ID, ACT_DTTM.

# Суть задачи в том, чтобы проверить подключения продуктов определенным пользователям, соединив файлы о продажах 
# с логами по подключениям в системе.

# Особенности данных:

#     сотрудники телемаркетинга не всегда указывают полный id, если 'id' нет в начале SUBS_ID, то нужно его добавить
#     поля в файлах могут быть расположены абсолютно случайным образом, но названия полей статичны
#     продажа не засчитывается, если отключение (END_DTTM) произошло меньше чем через 5 минут после подключения (START_DTTM)
#     если в файле с продажами встречается строка без указанного SUBS_ID, она пропускается

# Сохраните результат в датафрэйм с разделителем ;, содержащий корректные подключения.
# Note: обратите внимание на то, как pandas переводит дату из строки, возможно вам понадобится параметр format
```

```python
import pandas as pd
import datetime
```

```python
prod_activations = pd.read_csv('prod_activations_logs.csv', sep=';')
prod_activations
```

```python
tm_sales_1 = pd.read_csv('tm_sales_1.csv', sep=';')
tm_sales_1
```

```python
tm_sales_2 = pd.read_csv('tm_sales_2.csv', sep=';')
tm_sales_2
```

```python
tm_sales_3 = pd.read_csv('tm_sales_3.csv', sep=';')
tm_sales_3
```

```python
tm_sales = pd.concat([tm_sales_1, tm_sales_2, tm_sales_3])
tm_sales
```

```python
tm_sales.isna().sum()
```

```python
tm_sales = tm_sales.dropna(subset='SUBS_ID')
tm_sales
```

```python
def id_plus (name):
    if name.startswith('id'):
        new_name = name
    if not name.startswith('id'):
        new_name = 'id' + name
    return(new_name)
```

```python
tm_sales.SUBS_ID = tm_sales.SUBS_ID.apply(id_plus)
```

```python
tm_sales
```

```python
prod_activations.dtypes
```

```python
prod_activations.START_DTTM = pd.to_datetime(prod_activations.START_DTTM, format='%d-%m-%Y %H:%M')
```

```python
prod_activations.END_DTTM = pd.to_datetime(prod_activations.END_DTTM, format='%d-%m-%Y %H:%M')
```

```python
prod_activations.dtypes
```

```python
prod_activations['diff_time'] = prod_activations.END_DTTM - prod_activations.START_DTTM
prod_activations.sort_values('diff_time')
```

```python
prod_activations_end = prod_activations.drop(index=[4, 5])
prod_activations_end
```

```python
tm_sales
```

```python
prod_act_final = prod_activations_end.merge(tm_sales, how='inner', on=['SUBS_ID', 'PROD_ID']).sort_values('SUBS_ID')
prod_act_final
```

```python
prod_act_final.to_csv('prod_activations_end.csv', sep=';')
```
