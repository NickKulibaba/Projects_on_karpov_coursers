```python
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=6&q=cat')
cats = r.json()
```

```python
cats
```

```python
cat_ids_1000 = cats['objectIDs'][:1000]
len(cat_ids_1000)
```

```python
lst2 = []
df2 = pd.DataFrame()
cols2 = ['objectID', 'period']
```

```python
for num in cat_ids_1000:
    r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{num}')
    cat_ids = r_test.json()
    lst2.append([cat_ids['objectID'], cat_ids['period']])
    df2 = pd.DataFrame(lst2, columns=cols2)
df2
```

```python
df2 \
    .groupby('period') \
    .agg({'objectID': 'count'}) \
    .sort_values('objectID', ascending=False)
```
