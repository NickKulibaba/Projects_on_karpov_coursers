```python
#     Импортируйте библиотеку pandas как pd. Загрузите датасет bookings.csv с разделителем ;. 
#     Проверьте размер таблицы, типы переменных, а затем выведите первые 7 строк, чтобы посмотреть на данные. 
```

```python
import pandas as pd
```

```python
bookings = pd.read_csv('2_bookings.csv', sep=';')
bookings
```

```python
bookings.dtypes.sort_values()
```

```python
#     Приведите названия колонок к нижнему регистру и замените пробелы на знак нижнего подчеркивания.
```

```python
def lower_and_probel(name):
    new_name = name.lower().replace(' ','_')
    return new_name
```

```python
bookings = bookings.rename(columns=lower_and_probel)
bookings
```

```python
#     Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.
```

```python
bookings \
    .query('is_canceled ==0') \
    .groupby('country', as_index=False) \
    .agg({'is_canceled':'count'}) \
    .sort_values('is_canceled', ascending=False) \
    .head(5)
```

```python
#     На сколько ночей в среднем бронируют отели разных типов?
```

```python
bookings \
    .groupby('hotel', as_index=False) \
    .agg({'stays_total_nights': 'mean'}) \
    .round(2)
```

```python
#     Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально 
#     забронированного (reserved_room_type). Такое может произойти, например, по причине овербукинга. 
#     Сколько подобных наблюдений встретилось в датасете?
```

```python
bookings[['assigned_room_type','reserved_room_type']].query('assigned_room_type != reserved_room_type')
```

```python
#     Проанализируйте даты запланированного прибытия. – На какой месяц чаще всего успешно оформляли бронь в 2016? 
#     Изменился ли самый популярный месяц в 2017? 
```

```python
bookings \
    .query('is_canceled == 0') \
    .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
    .agg({'is_canceled': 'count'}) \
    .sort_values(['arrival_date_year', 'is_canceled'], ascending=False)
```

```python
bookings \
    .query('is_canceled == 0 and arrival_date_year == 2016') \
    .groupby('arrival_date_month', as_index=False) \
    .agg({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False) \
    .head(1)
```

```python
bookings \
    .query('is_canceled == 0 and arrival_date_year == 2017') \
    .groupby('arrival_date_month', as_index=False) \
    .agg({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False) \
    .head(1)
```

```python
#     Сгруппируйте данные по годам и проверьте, на какой месяц бронирования отеля 
#     типа City Hotel отменялись чаще всего в каждый из периодов
```

```python
bookings \
    .query("is_canceled == 1 and hotel == 'City Hotel'") \
    .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
    .agg({'is_canceled': 'count'}) \
    .sort_values(['arrival_date_year', 'is_canceled'], ascending=False)
```

```python
bookings \
    .query("is_canceled == 1 and hotel == 'City Hotel' and arrival_date_year == 2015") \
    .groupby('arrival_date_month', as_index=False) \
    .agg({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False) \
    .head(1)
```

```python
bookings \
    .query("is_canceled == 1 and hotel == 'City Hotel' and arrival_date_year == 2016") \
    .groupby('arrival_date_month', as_index=False) \
    .agg({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False) \
    .head(1)
```

```python
bookings \
    .query("is_canceled == 1 and hotel == 'City Hotel' and arrival_date_year == 2017") \
    .groupby('arrival_date_month', as_index=False) \
    .agg({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False) \
    .head(1)
```

```python
#     Посмотрите на числовые характеристики трёх переменных: adults, children и babies. 
#     Какая из них имеет наибольшее среднее значение?
```

```python
bookings[['adults', 'children', 'babies']].mean()
```

```python
#     Создайте колонку total_kids, объединив children и babies. Для отелей какого типа среднее 
#     значение переменной оказалось наибольшим?
```

```python
bookings['total_kids'] = bookings.children + bookings.babies
bookings
```

```python
bookings \
    .groupby('hotel', as_index=False) \
    .agg({'total_kids': 'mean'}) \
    .round(2)
```

```python
#     Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал 
#     хотя бы одного ребенка (total_kids), в противном случае – False. Посчитайте отношение количества 
#     ушедших пользователей к общему количеству клиентов, выраженное в процентах (churn rate). Укажите, среди 
#     какой группы показатель выше.
```

```python
bookings['has_kids'] = bookings.total_kids != 0
bookings
```

```python
all_with_kids = bookings.query('has_kids == True ').shape
all_with_kids
```

```python
canceled_with_kids = bookings.query('has_kids == True and is_canceled == 1').shape
canceled_with_kids
```

```python
all_no_kids = bookings.query('has_kids == False ').shape
all_no_kids
```

```python
canceled_no_kids = bookings.query('has_kids == False and is_canceled == 1').shape
canceled_no_kids
```

```python
churn_rate_with_kids = canceled_with_kids[0] / all_with_kids[0] * 100
churn_rate_with_kids
```

```python
churn_rate_no_kids = canceled_no_kids[0] / all_no_kids[0] * 100
churn_rate_no_kids
```
